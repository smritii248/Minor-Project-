```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)        
    text = re.sub(r"\[[0-9]+\]", "", text)       
    text = re.sub(r"\s+", " ", text)             
    return text.strip()

```

```python
def clean_example(example):
    example["page_title"] = clean_text(example["page_title"])
    example["page_text"] = clean_text(example["page_text"])
    return example

cleaned_dataset = dataset.map(clean_example)
```

```python
def filter_bad(example):
    return (
        example["page_title"] != "" and
        example["page_text"] != "" and
        len(example["page_text"]) > 30
    )

cleaned_dataset = cleaned_dataset.filter(filter_bad)
```
